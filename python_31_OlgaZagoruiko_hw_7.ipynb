```python
import os.path
import requests
import pandas as pd
```

```python
def download_document(file_name, document_url):
    if os.path.exists(file_name):
        pass
    else:
        response=requests.get(document_url)
        if response.status_code==200:
            with open(file_name,'wb') as file:
                file.write(response.content)
        else:
            print(f"Failed to download the document. Status code: {response.status_code}")


file_name='ipf_lifts.csv'
document_url='https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-10-08/ipf_lifts.csv'

download_document(file_name,document_url)
```

```python
df_ipf_lifts=pd.read_csv('ipf_lifts.csv')
display(df_ipf_lifts)
```

```python
df_lifts=df_ipf_lifts[df_ipf_lifts['division'].notnull()]
display(df_lifts)
```

2. Знайдіть рекорди кожної статі та кожного підрозділу в кожній із 3 вправ (жим лежачи, присідання та станова тяга).

```python
df_gender_and_division = df_lifts.groupby(by = ['sex', 'division']).agg({'best3squat_kg':'max','best3bench_kg':'max','best3deadlift_kg':'max'})
display(df_gender_and_division)
```

Інший спосіб

```python
df_gender_and_division2=df_lifts.loc[:,['sex','division','best3squat_kg','best3bench_kg','best3deadlift_kg']].groupby(by = ['sex', 'division']).max()
display(df_gender_and_division2)
```

3. Порахуйте кількість перемог кожного учасника, беручи за перемогу 1-е місце в конкретному дивізіоні

```python
df_winners=df_ipf_lifts[df_ipf_lifts['place']=='1']
display(df_winners)
```

```python
df_name_and_division = df_winners.groupby(by = ['name', 'division']).agg({'place':'count'})
print(df_name_and_division)
```

2 спосіб

```python
df_name_and_division2 = df_winners.groupby(by = ['name', 'division']).size()
print(df_name_and_division2)
```

3-й спосіб

```python
df_name_and_division3=df_winners[['name','division']].value_counts()
print(df_name_and_division3.head(20))
```

```python
df_new_winners=df_winners.loc[:,['name','sex','division']].merge(df_name_and_division3, how='left', on=['name','division'])
print(df_new_winners)
```

```python
df_group=df_new_winners.groupby(by=['division','sex'])['count'].transform('max')
print(df_group)
```
